```python
from sklearn.model_selection import train_test_split
n=500
x = np.float32(np.linspace(0,20,n))
np.random.seed(100)
error = np.random.standard_normal(n)
y = 3*x+error
y_true = 3*x 

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
linreg = LinearRegression()
linreg.fit(X_train.reshape(-1,1), y_train)
np.sqrt(metrics.mean_squared_error(y_test,linreg.predict(X_test.reshape(-1,1))))
```

```python
plt.scatter(y_test,linreg.predict(X_test.reshape(-1,1)))
```
